```python
def detecte_langage(message):
    languages_shared_words = {}
    words = wordpunct_tokenize(message)
    for language in stopwords.fileids():
        # stopwords pour chaque langue
        stopwords_liste = stopwords.words(language)
        # on retire les doublons
        words = set(words)
        # les mots communs entre stopwords 
        # d'une langue et les mots de message
        common_elements = words.intersection(stopwords_liste)
        # ajout du couple au dictionnaire
        languages_shared_words[language] = len(common_elements)
    # on retourne la langue avec le max de mots commun
    return  max(languages_shared_words, key = languages_shared_words.get)

def smileychoice():
    global mood
    if mood > 5 :
        return random.choice(smileyJoie)
    elif mood < -5 :
        return random.choice(smileyMechant)
    else :
        return random.choice(smiley)

def punctuation(message):
    for i in range(len(insulte)):
        if insulte[i] in message.lower() :
            return "agr"
    for i in range(len(positivity)) :
        if positivity[i] in message.lower() :
            return "gentil"
    
def detectIntention(message):
    global mood
    if message.isupper() or punctuation(message) == "agr" :
        mood -=1
    elif punctuation(message) == "gentil":
        mood += 1
        
```

```python
# Define replace_pronouns()
def replace_pronouns(message):

    message = message.lower()
    
    if ' me ' in message:
        # Replace 'me' with 'you'
        return re.sub('me','you',message)
    if 'my ' in message:
        # Replace 'my' with 'your'
        return re.sub('my','your',message)
    if ' your ' in message:
        # Replace 'your' with 'my'
        return re.sub('your','my',message)
    if ' you ' in message:
        # Replace 'you' with 'me'
        return re.sub('you','I',message)
    if " am "  in message : 
        return re.sub( "am","are",message)
    if " was "  in message: 
        return re.sub("was","were",message)
    if " I "  in message  : 
        return re.sub("I","you",message)
    if " I'd " in message : 
        return re.sub("I'd","you would",message)
    if " I've " in message : 
        return re.sub( "I've","you have",message)
    if " I'll "  in message:
        return re.sub("I'll", "you will",message)
    if " you've "in message: 
        return  re.sub("you've", "I have",message)
    if " you'll "in message: 
        mreturnre.sub( "you'll", "I will",message)
    if " yours " in message : 
        return re.sub("yours", "mine",message)
    if " I'm "  in message  : 
        return re.sub("I'm","you are",message)
    return message


# petite Fct qui s'occupe de la mémoire
def fctmemory(message):

    message = message.lower()
    if 'my' in message:
        tab = message.split("my")
        memory.append(tab[1])
        
    
# Define respond()
def respond(message):
    # Ajout de la mémoire
    fctmemory(message)
    response, phrase = match_rule(rules,message)
    if '{0}' in response:
      # Replace the pronouns in the phrase
      phrase = replace_pronouns(phrase)
      # Include the phrase in the response
      response = response.format(phrase)
    if mood < -10 :
        response = response.upper()
    return response

# Define a function that sends a message to the bot: send_message
def send_message(mess):
    global user_template
    # Print user_template including the user_message
    print(user_template.format(mess))
    # Get the bot's response to the message
    response = respond(mess)
    # Print the bot template including the bot's response.
    print(bot_template.format(response))
```
