## **ELIZA BASIQUE DATA CAMP**

```python
##Import
import re
import random
```

```python
#Les dictionnaires donnÃ©es par datacamp
rules = {
    'do you think (.*)': 
    [
        'if {0}? Absolutely.', 
        'No chance'
    ], 
    'do you remember (.*)': 
    [
        'Did you think I would forget {0}', 
        "Why haven't you been able to forget {0}", 
        'What about {0}', 'Yes .. and?'
    ], 
    'I want (.*)': 
    [
        'What would it mean if you got {0}', 
        'Why do you want {0}', 
        "What's stopping you from getting {0}"
    ], 
    'if (.*)': 
    [
        "Do you really think it's likely that {0}",
        'Do you wish that {0}', 
        'What do you think about {0}', 
        'Really--if {0}'
    ]
}
responses = {
    'statement': 
    [
        'tell me more!',
        'why do you think that?',
        'how long have you felt this way?',
        'I find that extremely interesting',
        'can you back that up?',
        'oh wow!',
        ':)'
    ],
    'question': 
    [
        "I don't know :(",
        'you tell me!'
    ]
}
```

```python
# Create templates
bot_template = "ELIZA : {0}"
user_template = "USER : {0}"
```

```python
# Define match_rule()
def match_rule(rules, message):
    response, phrase = "default", None
  
    # Iterate over the rules dictionary
    for pattern, rules_responses in rules.items():
        # Create a match object
        match = re.search(pattern, message)
        if match is not None:
            # Choose a random response
            response = random.choice(rules_responses)
            #print(response)
            if '{0}' in response:
                phrase = match.group(1)
                #print("phrase : " , phrase)
    if response == "default":
        if message.endswith("?"):
            # Return a random question
            response = random.choice(responses["question"])
        # Return a random statement
        response = random.choice(responses["statement"])
    # Return the response and phrase
    return response, phrase

# Define replace_pronouns()
def replace_pronouns(message):

    message = message.lower()
    if 'me' in message:
        # Replace 'me' with 'you'
        return re.sub('me','you',message)
    if 'my' in message:
        # Replace 'my' with 'your'
        return re.sub('my','your',message)
    if 'your' in message:
        # Replace 'your' with 'my'
        return re.sub('your','my',message)
    if 'you' in message:
        # Replace 'you' with 'me'
        return re.sub('you','I',message)

    return message

# Define respond()
def respond(message):
    # Call match_rule
    response, phrase = match_rule(rules,message)
    if '{0}' in response:
      # Replace the pronouns in the phrase
      phrase = replace_pronouns(phrase)
      # Include the phrase in the response
      response = response.format(phrase)
    return response


# Define a function that sends a message to the bot: send_message
def send_message(mess):
    global user_template
    # Print user_template including the user_message
    print(user_template.format(mess))
    # Get the bot's response to the message
    response = respond(mess)
    # Print the bot template including the bot's response.
    print(bot_template.format(response))
```

```python
# Send the messages
send_message("do you remember my last birthday")
send_message("do you think humans should be worried about AI")
send_message("I want a robot friend")
send_message("what if you could be anything you wanted")
send_message("test")
send_message("test ?")
```
